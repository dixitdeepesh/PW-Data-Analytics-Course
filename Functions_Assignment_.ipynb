## 1. What is the difference between a function and a method in Python?
- In Python, both functions and methods are reusable blocks of code designed to perform specific tasks, but they differ in usage and association. A function is an independent piece of code not tied to any object, defined using the `def` keyword and callable anywhere in a program. A method, however, is a function associated with an object or class, invoked using dot notation (e.g., `object.method()`).

  **Example:**
```
# Function
def greet():
    print("Hello!")
greet()
# Method (associated with a string object)
name = "Python"
print(name.lower())
```
The key distinction is that methods interact with the object they belong to, while functions are standalone.

## 2. Explain the concept of function arguments and parameters in Python.
- Parameters are variable names listed in a function's definition, acting as placeholders for values. Arguments are the actual values passed to the function during a call. Python supports:
 - Positional parameters (matched by position)
 - Keyword parameters (matched by name)
 - Default parameters (with default values)
 - Variable-length parameters (`*args`, `**kwargs`)
  
  **Example:**
```
def greet(name, message="Hello"):
    print(f"{message}, {name}!")
greet("Deepesh")                 # Uses default message
greet("Deepesh", "Good morning")  # Custom message
```
Parameters and arguments enhance function flexibility and reusability.

## 3. What are the different ways to define and call a function in Python?
- Functions can be defined using:
 - `def` keyword (standard functions)
 - `lambda` keyword (anonymous functions)
 - Recursive functions (self-calling)
 - Nested functions (defined inside other functions)

 **Example:**
```
# Regular function
def add(a, b):
    return a + b
# Lambda function
square = lambda x: x * x
# Calling the functions
print(add(5, 3))  # Output: 8
print(square(4))  # Output: 16
```
Functions improve code modularity, readability, and reusability.

## 4. What is the purpose of the return statement in a Python function?
- The `return` statement ends a function's execution and sends a result back to the caller. If omitted, the function returns `None`. It enables functions to produce usable outputs for further computation.
  
  **Example:**
```
def multiply(a, b):
    return a * b
result = multiply(4, 5)
print(result)  # Output: 20
```
`return` enhances control flow and result sharing in programs.

## 5. What are iterators in Python and how do they differ from iterables?
- An iterable is an object capable of returning elements one at a time (e.g., lists, tuples). An iterator is a specialized object that tracks its position during iteration, using `next()` to fetch values. Iterables can be converted to iterators with `iter()`.

 **Example:**
```
nums = [1, 2, 3]
it = iter(nums)
print(next(it))  # 1
print(next(it))  # 2
```

 Iterators are memory-efficient for large datasets, as they yield values on demand.

## 6. Explain the concept of generators in Python and how they are defined.
- Generators are iterators created using functions with the `yield` keyword. They produce values one at a time, pausing and resuming execution, which makes them memory-efficient.

 **Example:**
```
def countdown(n):
    while n > 0:
        yield n
        n -= 1
for number in countdown(3):
    print(number)  # Output: 3, 2, 1
```
Generators simplify iterator creation without manual state management.

## 7. What are the advantages of using generators over regular functions?
- Generators provide:
 - **Memory efficiency**: Generate values on demand, not storing entire sequences.
 - **Lazy evaluation**: Compute values only when needed.
 - **Simpler syntax**: Use `yield` to avoid complex iterator classes.

 **Example:**
```
def squares(limit):
    for i in range(limit):
        yield i * i
for val in squares(5):
    print(val)  # Output: 0, 1, 4, 9, 16
```
Generators are ideal for streaming data or large datasets.

## 8. What is a lambda function in Python and when is it typically used?
- A lambda function is an anonymous, single-expression function defined with the `lambda` keyword. Used for simple operations, often in `map()`, `filter()`, or `sorted()`.  
 **Example:**
```
add = lambda a, b: a + b
print(add(3, 4))  # Output: 7
```
Lambda functions are concise for temporary, inline use.

## 9. Explain the purpose and usage of the map() function in Python.
- `map()` applies a function to each item in an iterable, returning an iterator of results. It simplifies transformations without explicit loops.

 **Example:**
```
nums = [1, 2, 3]
doubles = map(lambda x: x * 2, nums)
print(list(doubles))  # Output: [2, 4, 6]
```
`map()` is ideal for functional-style data transformations.

## 10. What is the difference between map(), reduce(), and filter() functions in Python?
- **map()**: Applies a function to each item in an iterable.
- **filter()**: Selects items based on a condition (`True`/`False`).
- **reduce()**: Cumulatively applies a function to reduce an iterable to a single value.

 **Example:**
```
from functools import reduce
nums = [1, 2, 3, 4]
# map
print(list(map(lambda x: x * 2, nums)))  # [2, 4, 6, 8]
# filter
print(list(filter(lambda x: x % 2 == 0, nums)))  # [2, 4]
# reduce
print(reduce(lambda a, b: a + b, nums))  # 10
```
These functions enable concise, functional programming patterns.

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


input_numbers = input("Enter numbers separated by spaces: ")
number_list = list(map(int, input_numbers.split()))


result = sum_of_even_numbers(number_list)
print("Sum of even numbers:", result)
````

````python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(number):
    if number < 2 or number > 200:
        return "Please enter a number between 1 and 200."
    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return f"{number} is not a prime number."
    return f"{number} is a prime number."



num = int(input("Enter a number between 1 and 200: "))


print(is_prime(num))
````

````python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.term_count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.term_count >= self.max_terms:
            raise StopIteration
        if self.term_count == 0:
            self.term_count += 1
            return self.a
        elif self.term_count == 1:
            self.term_count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.term_count += 1
            return self.b


num_terms = int(input("Enter the number of Fibonacci terms to generate: "))
fib_iterator = FibonacciIterator(num_terms)

print("Fibonacci sequence:")
for num in fib_iterator:
    print(num, end=" ")
````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent


max_exp = int(input("Enter the maximum exponent: "))

print(f"Powers of 2 up to 2^{max_exp}:")
for value in powers_of_two(max_exp):
    print(value)
````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string.


with open("testfile.txt", "w") as f:
    f.write("Hello there\n")
    f.write("This is a test file\n")
    f.write("End of file\n")


def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()


file_path = "testfile.txt"  # Directly specify the file created above

print("File contents line by line:")
for line in read_file_line_by_line(file_path):
    print(line)

````

````python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.


data = [(1, 5), (2, 3), (4, 8), (3, 1)]


sorted_data = sorted(data, key=lambda x: x[1])


print("Sorted list of tuples (by second element):", sorted_data)
````

````python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.



input_temps = input("Enter temperatures in Degree Celsius (separated by spaces): ")


celsius_temps = list(map(float, input_temps.split()))


def c_to_f(temp):
    return (temp * 9/5) + 32


fahrenheit_temps = list(map(c_to_f, celsius_temps))

# Displaying the results
print("Temperatures in Celsius:", celsius_temps)
print("Temperatures in Fahrenheit:", fahrenheit_temps)
````

````python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.


def is_not_vowel(char):
    return char.lower() not in 'aeiou'


input_text = input("Enter a string: ")


filtered_text = ''.join(filter(is_not_vowel, input_text))


print("String without vowels:", filtered_text)
````
