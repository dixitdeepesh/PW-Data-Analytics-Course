### Q1: What is statistics, and why is it important?
  - Statistics is the discipline that deals with the collection, analysis, interpretation, and presentation of data. It enables decision-making by uncovering patterns, trends, and relationships in data.  
  Importance:
- Assists in decision-making
- Supports research and experiment design
- Aids in quality control and forecasting
- Used in various fields like business, government, healthcare, education, etc.


### Q2: What are the two main types of statistics?

  1. **Descriptive Statistics**: Summarizes data using measures like mean, median, and graphical representations.  
  2. **Inferential Statistics**: Makes predictions or generalizations about a population based on a sample.


### Q3: What are descriptive statistics?

- Descriptive statistics organize and present data using numbers, tables, and graphs to describe its basic features.  
Examples:  
 - Measures of central tendency
 - Measures of variability
 - Frequency distributions
 - Charts and plots


### Q4: What is inferential statistics?

- Inferential statistics involves drawing conclusions or inferences about a population using data collected from a sample.  
Examples:
 - Hypothesis testing
 - Confidence intervals
 - Regression analysis
 - ANOVA


### Q5: What is sampling in statistics?

- Sampling is selecting a subset from a population for measurement and analysis, used when studying the entire population is impractical.


### Q6: What are the different types of sampling methods?

- Probability Sampling:
 - Simple Random
 - Stratified
 - Cluster
 - Systematic

- Non-Probability Sampling:
 - Convenience
 - Purposive
 - Snowball
 - Quota


### Q7: What is the difference between random and non-random sampling?

 - **Random Sampling**: Each member has an equal chance of selection.
 - **Non-Random Sampling**: Selection based on researcher judgment or convenience.


### Q8: Define and give examples of qualitative and quantitative data.

 - **Qualitative Data**: Descriptive, categorical data. Examples: Gender, colors, blood group.
 - **Quantitative Data**: Numeric data. Examples: Height, weight, temperature.


### Q9: What are the different types of data in statistics?

 - Qualitative (Categorical): Nominal, Ordinal
 - Quantitative (Numerical): Discrete, Continuous


### Q10: Explain nominal, ordinal, interval, and ratio levels of measurement.

- Nominal: Categories without order (e.g., eye color)
- Ordinal: Categories with order (e.g., rankings)
- Interval: Numeric scales with equal intervals, no true zero (e.g., temperature)
- Ratio: Numeric scales with true zero (e.g., income)


### Q11: What is the measure of central tendency?

- A single value summarizing the center of a data set.
Types: Mean, Median, Mode


### Q12: Define mean, median, and mode.

- Mean: Sum of values divided by count.
- Median: Middle value when data is ordered.
- Mode: Most frequently occurring value.


### Q13: What is the significance of the measure of central tendency?

- Summarizes the dataset
- Aids comparisons
- Describes the distribution's center


### Q14: What is variance, and how is it calculated?

- Variance measures the average squared deviation from the mean.
Formula: Variance = sum((xi - mean)^2) / n


### Q15: What is standard deviation, and why is it important?

- Standard deviation is the square root of variance.  
Importance:
 - Indicates data spread
 - Used in quality control, forecasting, risk analysis


### Q16: Define and explain the term range in statistics.

- Range = Maximum value - Minimum value  
Measures total spread in data.


### Q17: What is the difference between variance and standard deviation?

- Variance: Average of squared deviations.
- Standard Deviation: Square root of variance, interpretable in data units.


### Q18: What is skewness in a dataset?

- Skewness measures the asymmetry of a distribution.
 - Positive Skew: Tail on the right
 - Negative Skew: Tail on the left


### Q19: What does it mean if a dataset is positively or negatively skewed?

- Positively Skewed: Mean > Median > Mode
- Negatively Skewed: Mean < Median < Mode


### Q20: Define and explain kurtosis.

- Kurtosis measures distribution tail thickness compared to a normal distribution.
 - Leptokurtic: Heavy tails
 - Mesokurtic: Normal
 - Platykurtic: Light tails


### Q21: What is the purpose of covariance?

- Covariance indicates the direction of relationship between two variables.
 - Positive: Move together
 - Negative: Move oppositely


### Q22: What does correlation measure in statistics?

- Correlation measures both the direction and strength of a linear relationship.
Ranges from -1 to +1.


### Q23: What is the difference between covariance and correlation?
  - **Covariance**  
  Direction only  
  Unit-dependent  
  Hard to interpret
    
- **Correlation**  
  Direction + Strength  
  Unit-free  
  Easy to interpret |

```python
#1.	How do you calculate the mean, median, and mode of a dataset?
from statistics import mean, median
from collections import Counter

try:
    # User-defined input
    data = list(map(float, input("Enter numbers separated by spaces: ").split()))

    if len(data) == 0:
        print("No numbers entered.")
    else:
        # Mean
        mean_value = mean(data)
        print("Mean:", mean_value)

        # Median
        median_value = median(data)
        print("Median:", median_value)

        # Mode (handling multiple modes)
        frequency = Counter(data)
        max_count = max(frequency.values())

        # Find all values with highest frequency
        modes = [key for key, count in frequency.items() if count == max_count]

        if max_count == 1:
            print("No mode found (all values appear only once).")
        else:
            print("Mode(s):", modes)

except ValueError:
    print("Invalid input! Please enter only numeric values separated by spaces.")

```

```python
#4.	Implement sampling techniques like random sampling and stratified sampling.
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit

try:

    n = int(input("Enter the number of data entries: "))

    if n <= 0:
        print("Number of entries must be positive.")
    else:
        # Collect data
        names, scores, grades = [], [], []

        for i in range(n):
            print(f"\nRecord {i+1}:")
            name = input("Enter Name: ")
            score = float(input("Enter Score (numeric): "))
            grade = input("Enter Grade (e.g., A, B, C): ")
            names.append(name)
            scores.append(score)
            grades.append(grade)


        df = pd.DataFrame({
            'Name': names,
            'Score': scores,
            'Grade': grades
        })

        print("\n--- Complete Dataset ---")
        print(df)

        # Random Sampling
        sample_size = int(input("\nEnter number of random samples to select: "))

        if sample_size > n:
            print("Sample size cannot be greater than total entries.")
        else:
            random_sample = df.sample(n=sample_size, random_state=1)
            print(f"\n--- Random Sampling ({sample_size} records) ---")
            print(random_sample)

            splitter = StratifiedShuffleSplit(n_splits=1, test_size=0.4, random_state=1)

            for train_index, test_index in splitter.split(df, df['Grade']):
                stratified_sample = df.loc[test_index]

            print(f"\n--- Stratified Sampling (40% based on 'Grade') ---")
            print(stratified_sample)

except ValueError:
    print("Invalid input! Please enter correct numeric and text values where required.")
```

```python
#5.	Write a Python function to calculate the range of a dataset.
def calculate_range(data):
    try:
        if len(data) == 0:
            return "Dataset is empty. Cannot calculate range."
        else:
            data = list(map(float, data))  # ensure all values are numeric
            range_value = max(data) - min(data)
            return range_value
    except ValueError:
        return "Invalid input! All values must be numeric."

# User-defined input
user_input = input("Enter numbers separated by spaces: ").split()

# Call function and display result
result = calculate_range(user_input)
print("Range of the dataset:", result)
```

```python
#6.	Create a dataset and plot its histogram to visualize skewness.
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

try:
    # User-defined input
    data = list(map(float, input("Enter numbers separated by spaces: ").split()))

    if len(data) == 0:
        print("No data entered.")
    else:
        # Calculate skewness
        skewness_value = skew(data)

        # Display skewness value
        print("Skewness of the dataset:", skewness_value)

        # Plot histogram using seaborn
        plt.figure(figsize=(8, 5))
        sns.histplot(data, kde=True, bins=8, color='skyblue')
        plt.title("Histogram to Visualize Skewness")
        plt.xlabel("Values")
        plt.ylabel("Frequency")
        plt.grid(True)
        plt.show()

        # Interpret skewness numerically
        if skewness_value > 0:
            print("The distribution is positively skewed (tail on the right).")
        elif skewness_value < 0:
            print("The distribution is negatively skewed (tail on the left).")
        else:
            print("The distribution is approximately symmetric.")

except ValueError:
    print("Invalid input! Please enter only numeric values separated by spaces.")
```

```python
#7.	Calculate skewness and kurtosis of a dataset using Python libraries.
from scipy.stats import skew, kurtosis

try:
    # User-defined input
    data = list(map(float, input("Enter numbers separated by spaces: ").split()))

    if len(data) == 0:
        print("No data entered.")
    else:
        # Calculate Skewness
        skewness_value = skew(data)
        print("Skewness:", skewness_value)

        # Interpret Skewness
        if skewness_value > 0:
            print("The distribution is positively skewed (tail on the right).")
        elif skewness_value < 0:
            print("The distribution is negatively skewed (tail on the left).")
        else:
            print("The distribution is approximately symmetric.")

        # Calculate Kurtosis
        kurtosis_value = kurtosis(data)
        print("Kurtosis:", kurtosis_value)

        # Interpret Kurtosis
        if kurtosis_value > 0:
            print("Leptokurtic distribution (heavy tails).")
        elif kurtosis_value < 0:
            print("Platykurtic distribution (light tails).")
        else:
            print("Mesokurtic distribution (normal-like tails).")

except ValueError:
    print("Invalid input! Please enter only numeric values separated by spaces.")
```

```python
#8.	Generate a dataset and demonstrate positive and negative skewness
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Generate synthetic datasets
np.random.seed(1)

# Positive skew: Exponential distribution
positive_skew_data = np.random.exponential(scale=2, size=1000)

# Negative skew: Exponential reversed
negative_skew_data = -np.random.exponential(scale=2, size=1000)

# Symmetric distribution: Normal
symmetric_data = np.random.normal(loc=0, scale=1, size=1000)

# Function to plot and print skewness
def plot_and_skew(data, title):
    skew_value = skew(data)
    print(f"{title} - Skewness: {skew_value:.3f}")

    plt.figure(figsize=(7, 4))
    sns.histplot(data, kde=True, bins=30, color='skyblue')
    plt.title(f"{title}\nSkewness: {skew_value:.3f}")
    plt.xlabel("Values")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

# Positive skew
plot_and_skew(positive_skew_data, "Positively Skewed Distribution")

# Negative skew
plot_and_skew(negative_skew_data, "Negatively Skewed Distribution")

# Symmetric (for comparison)
plot_and_skew(symmetric_data, "Symmetric (Normal) Distribution")
```

```python
#9.	Write a Python script to calculate covariance between two datasets.
import numpy as np

try:
    # User-defined input for dataset X
    data_x = list(map(float, input("Enter numbers for Dataset X (space-separated): ").split()))

    # User-defined input for dataset Y
    data_y = list(map(float, input("Enter numbers for Dataset Y (space-separated): ").split()))

    # Check both datasets are non-empty and have same length
    if len(data_x) == 0 or len(data_y) == 0:
        print("One or both datasets are empty.")
    elif len(data_x) != len(data_y):
        print("Datasets must have the same number of elements.")
    else:
        # Calculate covariance matrix
        covariance_matrix = np.cov(data_x, data_y)
        covariance_value = covariance_matrix[0, 1]

        # Display results
        print("\nCovariance Matrix:\n", covariance_matrix)
        print("\nCovariance between X and Y:", covariance_value)

except ValueError:
    print("Invalid input! Please enter only numeric values separated by spaces.")
```

```python
#10.	Write a Python script to calculate the correlation coefficient between two datasets.
import numpy as np

try:
    # User-defined input for dataset X
    data_x = list(map(float, input("Enter numbers for Dataset X (space-separated): ").split()))

    # User-defined input for dataset Y
    data_y = list(map(float, input("Enter numbers for Dataset Y (space-separated): ").split()))

    # Check both datasets are non-empty and have same length
    if len(data_x) == 0 or len(data_y) == 0:
        print("One or both datasets are empty.")
    elif len(data_x) != len(data_y):
        print("Datasets must have the same number of elements.")
    else:
        # Calculate correlation coefficient matrix
        correlation_matrix = np.corrcoef(data_x, data_y)
        correlation_value = correlation_matrix[0, 1]

        # Display results
        print("\nCorrelation Coefficient Matrix:\n", correlation_matrix)
        print("\nCorrelation Coefficient between X and Y:", correlation_value)

        # Interpretation
        if correlation_value > 0:
            print("Positive correlation.")
        elif correlation_value < 0:
            print("Negative correlation.")
        else:
            print("No correlation.")

except ValueError:
    print("Invalid input! Please enter only numeric values separated by spaces.")

```

```python
#12.	Implement and compare simple random sampling and systematic sampling.
import numpy as np
import pandas as pd

try:
    data = list(map(float, input("Enter dataset values (space-separated): ").split()))
    n = int(input("Enter sample size: "))

    if n > len(data) or n <= 0:
        print("Sample size must be positive and less than or equal to dataset size.")
    else:
        # Simple Random Sampling
        random_sample = np.random.choice(data, size=n, replace=False)
        print("Simple Random Sample:", random_sample)

        # Systematic Sampling
        k = len(data) // n
        start = np.random.randint(0, k)
        systematic_sample = [data[i] for i in range(start, len(data), k)][:n]
        print("Systematic Sample:", systematic_sample)

except ValueError:
    print("Invalid input! Enter numbers separated by spaces.")


```

```python
#13.	Calculate the mean, median, and mode of grouped data.
from statistics import mean, median, mode

try:
    intervals = list(map(float, input("Enter class mid-points (space-separated): ").split()))
    frequencies = list(map(int, input("Enter corresponding frequencies (space-separated): ").split()))

    if len(intervals) != len(frequencies):
        print("Class mid-points and frequencies must be of equal length.")
    else:
        data = []
        for mid, freq in zip(intervals, frequencies):
            data.extend([mid] * freq)

        print("Mean:", mean(data))
        print("Median:", median(data))
        try:
            print("Mode:", mode(data))
        except:
            print("No unique mode found.")

except ValueError:
    print("Invalid input.")

```

```python
#18.	Implement Z-score normalization and explain its significance.
import numpy as np

data = list(map(float, input("Enter dataset values (space-separated): ").split()))
mean_val = np.mean(data)
std_val = np.std(data)

z_scores = [(x - mean_val) / std_val for x in data]

print("Z-score Normalized Data:", z_scores)
print("""
Significance:
Z-score normalization rescales data so the mean is 0 and standard deviation is 1.
It helps in comparing variables of different scales and detecting outliers effectively.
""")
```
